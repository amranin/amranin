### Problem 9: Matplotlib (10 points)
Let's say the worth of a house in an area gets halved every year. For example, a house worth 1 million dollars today is valued at 0.5 million after 1 year, 0.25 million after 2 years and so on. This is called exponential decay. Given how much a house is valued today, we want to plot how much it will be worth after 0, 1, 2, ..., 10 years.

````
import math

def future_worth(t):
    '''
    Returns worth after t years
    At t = 0, worth = 700,000
    '''
    base_worth = 700000
    return base_worth * math.pow(2, -t)
````

````
import matplotlib.pyplot as plt
import numpy as np

# REPLACE _ WITH APPROPRIATE QUANTITY
# hint: np.arange(n) is equivalent to [0,1,2,..,n-1] - remember starts w 0
x = np.arange(11)

y = []
for t in x:
    # adds new entry to y in order
    y.append(future_worth(t))

# COMPLETE THE PLOT

plt.title('House worth vs time')

## COMPLETE CODE BELOW
plt.plot(x,y, color='pink')
plt.xlabel('Time in Years')
plt.ylabel('House Worth in USD')
plt.show()
````
